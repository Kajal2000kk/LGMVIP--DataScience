## Step 4: Data Visualization 

### Countries Affected

```python
plt.subplots(figsize=(10,10))
sns.barplot(df['Country'].value_counts()[:10].index,df['Country'].value_counts()[:10])
plt.title('Top Countries Affected',fontsize=20)
plt.xlabel('Countries')
plt.ylabel('Count')
plt.xticks(rotation = 50)
plt.show()
```

### Number of killed 

```python
df[['Attacktype','Killed']].groupby(["Attacktype"],axis=0).sum().plot(kind='bar',figsize=(5,5))
plt.xticks(rotation=50)
plt.title("Number of killeds ",fontsize=20)
plt.ylabel('Number of peoples')
plt.xlabel('Attack types')
plt.show()
```

### Number of Wounded 

```python
df[['Attacktype','Wounded']].groupby(["Attacktype"],axis=0).sum().plot(kind='bar',figsize=(5,5),color='red')
plt.xticks(rotation=50)
plt.title("Number of Wounded ",fontsize=20)
plt.ylabel('Number of peoples')
plt.xlabel('Attack types')
plt.show()
```

### Terrorist Activities By Region In Each Year

```python
pd.crosstab(df['Year'], df['Region']).plot(kind = 'area', stacked = False, figsize = (10,5))
plt.title('Terrorist Activities By Region In Each Year', fontsize = 25)
plt.ylabel('Number of Attacks', fontsize = 20)
plt.xlabel("Year", fontsize = 20)
plt.show()
```

```python
Attack_per_country = df['Country'].value_counts()[:10]
print(Attack_per_country)
```

### Method of Attack

```python
plt.figure(figsize=(10,3))
sns.countplot(x = df['Attacktype'], order = df['Attacktype'].value_counts().index)
plt.xticks(rotation=45)
plt.xlabel('Method')
plt.title('Method of Attack',fontsize=20)
plt.show()
```

### Counting terrorist activities each year through countplot 

```python
plt.subplots(figsize=(15,5))
sns.countplot('Year',data = df)
plt.xticks(rotation=90)
plt.title('Number Of Terrorist Activities Each Year')
plt.grid()
plt.show()
```
